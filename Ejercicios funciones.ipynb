## Funciones - Ejercicios

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

Para todos los ejercicios utilizar un rango del -5 al 5.1 en intervalos de 0.1 (usar función np.arange)

1. Graficar la función f(x) = 4x + 2

```python
vector_x = [i for i in range(-10,11)]
vector_y = [4*x + 2 for x in vector_x]
print(vector_x)
print(vector_y)

plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
sns.scatterplot(x = vector_x, y = vector_y, color = "red")
plt.show()
```

2. Graficar la función f(x) = -x^2 + 1

```python
vector_x = [i for i in np.arange(-5, 5, 0.1)]
vector_y = [-x ** 2 + 1 for x in vector_x]

plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
sns.scatterplot(x = vector_x, y = vector_y, color = "red")
plt.show()
```

3. Graficar la función f(x) = 4x^3 - 2x^2 + 1

```python
vector_x = [i for i in np.arange(-5, 5, 0.1)]
vector_y = [4 * x ** 3 - 2 * x ** 2 + 1 for x in vector_x]

plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
sns.scatterplot(x = vector_x, y = vector_y, color = "red")
plt.show()
```

4. Graficar la función f(x) = 2^x - 3

```python
vector_x = [i for i in np.arange(-5, 5, 0.1)]
vector_y = [2 ** x - 3 for x in vector_x]

plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
sns.scatterplot(x = vector_x, y = vector_y, color = "red")
plt.show()
```

5. Graficar la función f(x) = x^6 - 3x^2

```python
vector_x = [i for i in np.arange(-5, 5, 0.1)]
vector_y = [(x ** 6) - (3 * x ** 2) for x in vector_x]

plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
sns.scatterplot(x = vector_x, y = vector_y, color = "red")
plt.show()
```

6. Graficar la siguiente función
$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

```python
vector_x = [i for i in np.arange(-5, 5, 0.1)]
vector_y = [1/(math.sqrt(2*math.pi)) * math.exp(-x**2/2) for x in vector_x]
plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
sns.scatterplot(x = vector_x, y = vector_y, color = "red")
plt.show()
```

```python
lam = 1
vector_x = [i for i in np.arange(0, 6, 0.01)]
vector_y = [lam * math.exp(-lam*x) for x in vector_x]
plt.title("Densidad de probabilidad Exponencial")
plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
plt.show()
```

```python
lam = 1
vector_x = [i for i in np.arange(0, 6, 0.01)]
vector_y = [1 - math.exp(-lam*x) for x in vector_x]
plt.title("Distribución de probabilidad Exponencial")
plt.axvline(x=0, color = "gray", ls = "--")
plt.axhline(y=0, color = "gray", ls = "--")
sns.lineplot(x = vector_x, y = vector_y)
plt.show()
```
