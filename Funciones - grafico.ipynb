## Gráfico de funciones

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
````

### Función lineal
Grafico la función f(x) = 3x + 2

````python
# genero un vector de números y los llamo x
vector_x = [i for i in range(-10,10)]
vector_y = [3 * x + 2 for x in vector_x]
````

````python
plt.axvline(x=0, color='gray', linestyle='--',alpha = 0.5)
plt.axhline(y=0, color='gray', linestyle='--',alpha = 0.5)
sns.lineplot(x=vector_x,y=vector_y)
````

### Función Exponencial
Calcular cuantas veces hay que doblar un papel para que tenga la altura de la distancia de la luna a la tierra (400.000 km) si el espesor del papel es de 0.1mm. Graficar crecimiento de altura en cada doblez.

En cada doblez se duplica la altura, por lo que podríamos expresar la función de crecimiento como:
```
f(x) = 0.1mm * 2^x
f(x) = 0.0001m * 2^x = 10^-4 m * 2^x
f(x) = 0.0000001km * 2^x = 10^-7 km * 2^x
```

````python
# puedo despejar la ecuación cuando f(x) sea igual a 400.000, pero vamos a iterar y ver en cuantos dobleces llegamos
vector_x = [i for i in range(43)]
vector_y = [10**-7 * 2**x for x in vector_x]
````

````python
print(vector_x)
print(vector_y)
````

````python
sns.lineplot(x=vector_x,y=vector_y)
````

### Circunferencia
Graficar la función x^2 + y^2 = 100

Despejando sería y= +- sqrt(10-x^2)

````python
# genero un vector de números y los llamo x. La circunferencia es de radio 10, por lo que genero números del -10 al 10
vector_x = [i for i in np.arange(-10,10.1,0.1)]
vector_y_pos = [math.sqrt(100-i**2) for i in vector_x]
vector_y_neg = [-math.sqrt(100-i**2) for i in  vector_x]
````

````python
print(vector_x[:5])
print(vector_y_pos[:5])
print(vector_y_neg[:5])
````

````python
sns.lineplot(x=vector_x,y=vector_y_pos)
sns.lineplot(x=vector_x,y=vector_y_neg)
````

### Corazón
graficar la función x^2 + (y - x^{2/3})^2 = 100

despejando: 

$$
y = x^{\frac{2}{3}} \pm \sqrt{100-x^2}
$$

````python
# genero un vector de números y los llamo x. La circunferencia es de radio 10, por lo que genero números del -10 al 10. 
# Tengo que hacer por un lado los negativos y por otro los positivos
vector_x_pos = [i for i in np.arange(0,10.1,0.1)]
vector_x_neg = [i for i in np.arange(-10,0.1,0.1)]

vector_y_pos = [i**(2/3) + math.sqrt(100-i**2) for i in vector_x_pos]
vector_y_neg = [i**(2/3) -math.sqrt(100-i**2) for i in  vector_x_pos]

sns.lineplot(x=vector_x,y=vector_y_pos)
sns.lineplot(x=vector_x,y=vector_y_neg)
sns.lineplot(x=vector_x_neg[::-1],y=vector_y_neg)
sns.lineplot(x=vector_x_neg[::-1],y=vector_y_pos)
````
